# Generate RED data

```python
import calculus as calc
import auxiliar as aux

########################

data     = 'path_data'


N_data   = 150         # number of runs we want to do
N        = 5           # number of sites
ncycles  = 8           # number of layers


max_iter = 10e6
total_t  = 200         # total evolution time

mode = 'pairs_conc'    # possibilities: 'evenodd','pairs','pairs_nnn','pairs_conc','individual'


counter = [50,80,120,200]    # limit counters of repeated energy
alphas  = [0.3,0.5,0.05,3]   # alphas used in each stage

#########################

#setting the initial ansatz
ansatz = calc.psi_ansatz(N)

# setting all the parameters
calc.parameters(N,ncycles,ansatz,mode,1,data)
# calculate the number of theta variables
num = aux.nVariables(N,ncycles,mode)  

for i in range(N_data):
    print('RED - Run: ',i,'/',N_data)
    aux.write('new-run','random_estimation_descent',data)
    
    #setting the theta with the corresponding seed
    t0=calc.theta_ansatz(N,8,mode,ncycles,seed=i)
    
    # do the descent using RED
    calc.descent_test(t0,max_iter,counter,alphas,total_t)
```

# Circuit tests

```python
import calculus as calc
import auxiliar as aux

########################

data     = '/home/juan/Desktop/data/'
filename = 'DG_modes'

N_data   = 150         # number of runs we want to do
N        = 5           # number of sites
ncycles  = 8           # number of layers

max_iter = 10e6
alpha    = 10e-1/2.7   # learning rate
total_t  = 200         # total evolution time

modes = ['evenodd','pairs_nnn','pairs_conc','individual']

#########################

#setting the initial ansatz
ansatz = calc.psi_ansatz(N)

# setting all the parameters
calc.parameters(N,ncycles,ansatz,mode,1,data)
# calculate the number of theta variables
num = aux.nVariables(N,ncycles,mode)  

# iterating over different modes
for mode in modes:
    
    # for each mode iterate the needed number of times
    for i in range(N_data):
        
        print('DG - Run: ',i,'/',N_data,' circuit ',mode)
        aux.write('new-run',filename,data)

        #setting the theta with the corresponding seed
        t0=calc.theta_ansatz(N,8,mode,ncycles,seed=None)

        # do the descent using DG
        calc.gradient_descent(calc.energy, t0, alpha, 0.001, max_iter, total_t,file=filename,write_mode=True) 
```
